```python
from collections import deque


JUG1 = 4
JUG2 = 3


GOAL = 2


visited = set()

def dfs(state, path):
    x, y = state

    
    if state in visited:
        return False

    visited.add(state)
    path.append(state)

  
    if x == GOAL or y == GOAL:
        return True

 

    possible_states = []

   
    possible_states.append((JUG1, y))


    possible_states.append((x, JUG2))

    possible_states.append((0, y))

   
    possible_states.append((x, 0))

   
    transfer = min(x, JUG2 - y)
    possible_states.append((x - transfer, y + transfer))

   
    transfer = min(y, JUG1 - x)
    possible_states.append((x + transfer, y - transfer))

    for new_state in possible_states:
        if dfs(new_state, path):
            return True

    path.pop()
    return False


path = []
dfs((0, 0), path)

print("DFS Solution Path:")
for step in path:
    print(step)
```
